```python
import numpy as np
import json
import pandas as pd
import matplotlib.pyplot as pyplot
import jmespath
```

```python
with open('cloudwatch/load_balancer/active_connection_count_metric.json', 'r') as f:
      data = json.load(f)
```

```python
active_connection_timestamps = list(reversed(['2022/10/16 18:34', '2022/10/16 18:32', '2022/10/16 18:07', '2022/10/16 18:00', '2022/10/16 17:54', '2022/10/16 17:53', '2022/10/16 17:52', '2022/10/16 17:51', '2022/10/16 17:50', '2022/10/16 17:49', '2022/10/16 17:48', '2022/10/16 17:47'])) 
active_connection_values = list(reversed([2.0, 2.0, 1.0, 2.0, 5.0, 4.0, 9.0, 10.0, 4.0, 8.0, 10.0, 16.0])) 
x_axis = np.arange(12)
```

```python
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        pyplot.text(i, y[i], y[i], ha = 'center')
```

```python
fig = pyplot.figure(figsize=(12,12))
#fig.suptitle("Active connection count")
ax = fig.add_subplot(111)
ax.bar(active_connection_timestamps, active_connection_values)
ax.set_xticklabels(active_connection_timestamps,rotation = 45)
addlabels(active_connection_timestamps, active_connection_values)
pyplot.xlabel("Timestamps")
pyplot.ylabel("# of active TCP Connection per times")
pyplot.plot(x_axis,active_connection_values, color="r")
pyplot.savefig("document/Active_Connection.pdf", dpi=400)
pyplot.show()
```

```python
active_connection_timestamps = list(reversed(['2022/10/16 17:57', '2022/10/16 17:56', '2022/10/16 17:55', '2022/10/16 17:54', '2022/10/16 17:53', '2022/10/16 17:52', '2022/10/16 17:51', '2022/10/16 17:50', '2022/10/16 17:49', '2022/10/16 17:48', '2022/10/16 17:47', '2022/10/16 17:46', '2022/10/16 17:45', '2022/10/16 17:44'])) 
active_connection_values = list(reversed([4.0, 4.0, 4.0, 4.0, 4.0, 4.0, 4.0, 4.0, 4.0, 3.0, 2.0, 1.0, 0.5, 0.5])) 
x_axis = np.arange(14)
```

```python
# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        pyplot.text(i, y[i], y[i], ha = 'center')
```

```python
fig = pyplot.figure(figsize=(12,12))
#fig.suptitle("Active connection count")
ax = fig.add_subplot(111)
ax.bar(active_connection_timestamps, active_connection_values)
ax.set_xticklabels(active_connection_timestamps,rotation = 45)
addlabels(active_connection_timestamps, active_connection_values)
pyplot.xlabel("Timestamps")
pyplot.ylabel("# of healty target")
pyplot.plot(x_axis,active_connection_values, color="r")
pyplot.savefig("document/Tgt_Grp1_Healthy.pdf", dpi=400)
pyplot.show()
```
